```python
#Need to compare the statistics by (same file, indivdual or subject) trials
#Ex: subject 1 - 3x trials
#.describe() - mean, std data

#Thigh, Knee, Calf Angle plots
# with Mode-Phase Algo
#Saved gait durations / plot in folder
 
import os
import pandas as pd
import json
import matplotlib.pyplot as plt
%matplotlib inline
import datetime



# ---- Set/select folder path for files ----
folder_path = r"C:\Users\KevWal\Downloads\20250515"  # ‚Üê Update if needed

# Load files
DataFrame = {}
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        full_path = os.path.join(folder_path, filename)

        data = []
        with open(full_path, "r") as file:
            for line in file:
                clean = line.strip().strip('"').replace('""', '"')
                try:
                    parsed = json.loads(clean)
                    data.append(parsed)
                except json.JSONDecodeError:
                    print(f"Error decoding line in {filename}")

        df = pd.json_normalize(data)

        if 'input' in df and 'output' in df:
            input_df = pd.DataFrame(df['input'].tolist()).add_prefix('input')
            output_df = pd.DataFrame(df['output'].tolist()).add_prefix('output')
            full_df = pd.concat([df[['sn', 'time']], input_df, output_df], axis=1)

            full_df.rename(columns={
                'sn': 'sample_number',
                'input0': 'mode',
                'input1': 'phase',
                'input2': 'flex_damp_measure',
                'input3': 'ext_damp_measure',
                'input4': 'knee_angle (degree)',
                'input5': 'knee_velocity(degree/s)',
                'input6': 'thigh_angle(degree)',
                'input7': 'thigh_velocity(degree/s)',
                'input8': 'calf_angle(degree)',
                'input9': 'calf_velocity(degree/s)',
                'input10': 'calf_ang_acc_smooth(degree/s¬≤)',
                'input11': 'acc_abs_mag',
                'input12': 'acc_vertical_world',
                'input13': 'abs_knee_vel_avg(degree/s)',
                'input14': 'knee_velocity_history',
                'input15': 'reserved',
                'output0': 'motor_flex',
                'output1': 'motor_extent',
                'time': 'timestamp'
            }, inplace=True)

            df_name = filename.replace('.txt', '')
            DataFrame[df_name] = full_df

# Filter for specific files
prefixes = ["14-03-58"] #["10-33-13", "14-02-29","13-30-18"]

# --- Utility Functions ---

def extract_subject(file_name):
    return file_name.split('_')[0]  # Adjust this depending on naming

def seconds_to_hms(seconds):
    return str(datetime.timedelta(seconds=seconds))
    
def compress_repeats_with_index(lst):
    if not lst:
        return [], []
    compressed, indices = [lst[0]], [0]
    for i in range(1, len(lst)):
        if lst[i] != compressed[-1]:
            compressed.append(lst[i])
            indices.append(i)
    return compressed, indices

def find_sequence_with_original_indices(column, sequence):
    matches = []
    compressed, indices = compress_repeats_with_index(column)
    for i in range(len(compressed) - len(sequence) + 1):
        if compressed[i:i + len(sequence)] == sequence:
            start = indices[i]
            j = indices[i + len(sequence) - 1]
            while j + 1 < len(column) and column[j + 1] == sequence[-1]:
                j += 1
            matches.append((start, j))
    return matches

def find_gait_transitions(column):
    GC_013 = find_sequence_with_original_indices(column, [0, 1, 3])
    GC_0123 = find_sequence_with_original_indices(column, [0, 1, 2, 3])
    matches = sorted(GC_013 + GC_0123, key=lambda x: x[0])

    segments = []
    prev_end = -1
    for start, end in matches:
        if prev_end + 1 < start:
            segments.append((prev_end + 1, start - 1, 'NonGait'))
        segments.append((start, end, 'Gait'))
        prev_end = end
    if prev_end + 1 < len(column):
        segments.append((prev_end + 1, len(column) - 1, 'NonGait'))
    return segments

def label_phases_within_cycles(column, gait_cycles):
    labels = ['Other'] * len(column)
    for start, end in gait_cycles:
        for i in range(start, end + 1):
            if column[i] in [0, 3]:
                labels[i] = 'Stance'
            elif column[i] in [1, 2]:
                labels[i] = 'Swing'
    return labels

def label_gait_and_nongait(column):
    labels = ['Other'] * len(column)
    segments = find_gait_transitions(column)
    for start, end, label in segments:
        for i in range(start, end + 1):
            labels[i] = label
    return labels

def label_gait_phases(df):
    phase_col = df['phase'].tolist()
    gait_segments = find_gait_transitions(phase_col)
    gait_cycles = [(s, e) for s, e, label in gait_segments if label == 'Gait']
    gait_labels = label_gait_and_nongait(phase_col)
    phase_labels = label_phases_within_cycles(phase_col, gait_cycles)
    final_labels = [p if g == 'Gait' else 'NonGait' for g, p in zip(gait_labels, phase_labels)]
    return gait_cycles, gait_labels, phase_labels, final_labels



# ---- Color and name mapping ----
phase_colors = {
    0: 'red',    # Stance
    1: 'blue',   # PreSwing
    2: 'green',  # Swing
    3: 'black'   # Weight Acceptance
}

phase_names = {
    0: 'Stance',
    1: 'PreSwing',
    2: 'Swing',
    3: 'Weight Acceptance'
}

# ---- Window configuration ----
window_seconds = 5
fs = 100
window_size = window_seconds * fs

summary_stats = []
# ---- Process each file ----
for file_name, df in DataFrame.items():
    if prefixes and not any(file_name.startswith(p) for p in prefixes):
        continue

    print(f"\nüìÅ Processing file: {file_name}")
    save_folder = os.path.join(folder_path, file_name.split("_")[0])
    os.makedirs(save_folder, exist_ok=True)

    filtered_df = df[df['mode'] == 0].reset_index(drop=True)

    # Usage on your timestamp column:
    filtered_df['human_time'] = filtered_df['timestamp'] - filtered_df['timestamp'].iloc[0]  # elapsed seconds
    filtered_df['human_time'] = filtered_df['human_time'].apply(seconds_to_hms)


    overall_duration = df['timestamp'].iloc[-1] - df['timestamp'].iloc[0]
    # Convert overall duration to minutes and seconds
    overall_mins, overall_secs = divmod(overall_duration, 60)
    minutes_duration = f"{int(overall_mins)} min {int(overall_secs)} sec"

    print(f"\nüïí Overall Sampling Duration: {overall_duration:.2f} seconds "
          f"({minutes_duration})\n")

    
    # ‚ö†Ô∏è Skip files with no mode 0 data
    if filtered_df.empty:
        print(f"‚ö†Ô∏è Skipping {file_name} ‚Äî no valid mode 0 data.")
        continue

     # ---- Segment and plot ----
    for start in range(0, len(filtered_df), window_size):
        end = min(start + window_size, len(filtered_df))
        segment_df = filtered_df.iloc[start:end]
    
        # Create a filename-friendly timestamp for saving
        segment_start_time = segment_df['timestamp'].iloc[0]
        timestamp_str = f"{segment_start_time:.2f}".replace('.', '_')
    
        # Gait and phase detection per segment
        gait_cycles, gait_labels, phase_labels, final_labels = label_gait_phases(segment_df)
    
    
        segment_df = segment_df.copy()
        segment_df['gait_label'] = gait_labels
        segment_df['phase_label'] = phase_labels
        segment_df['final_label'] = final_labels
    
    
        gait_df = segment_df[segment_df['gait_label'] == 'Gait'].reset_index(drop=True)
        if gait_df.empty:
            continue  # Skip if no gait data
    
        fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
        axs[0].set_title(f"{file_name} ‚Äî Gait Segment Start: {segment_start_time:.2f}s")
    
        start_idx = 0
        for i in range(1, len(gait_df)):
            if gait_df['phase'][i] != gait_df['phase'][i - 1]:
                sub = gait_df.iloc[start_idx:i]
                phase = gait_df['phase'][i - 1]
                color = phase_colors.get(phase, 'gray')
                label = phase_names.get(phase, f'Phase {phase}')
                legend_labels = axs[0].get_legend_handles_labels()[1]
    
                axs[0].plot(sub['timestamp'], sub['knee_angle (degree)'], color=color,
                    label=label if label not in legend_labels else "", linewidth=2, alpha=1)
              
    
                axs[1].plot(sub['timestamp'], sub['thigh_angle(degree)'], color=color, linewidth=2, alpha=1)
                
                # axs[2].plot(sub['timestamp'], sub['calf_angle(degree)'], color=color, linewidth=2, alpha=1)
                
                # axs[3].plot(sub['timestamp'], sub['abs_knee_vel_avg(degree/s)'], color=color, linewidth=2, alpha=1)
    
    
                start_idx = i
    
        # Final segment
        sub = gait_df.iloc[start_idx:]
        if not sub.empty:
            phase = sub['phase'].iloc[-1]
            color = phase_colors.get(phase, 'gray')
            label = phase_names.get(phase, f'Phase {phase}')
            legend_labels = axs[0].get_legend_handles_labels()[1]
        
            axs[0].plot(sub['timestamp'], sub['knee_angle (degree)'], color=color, linewidth=2, alpha=1)
            axs[1].plot(sub['timestamp'], sub['thigh_angle(degree)'], color=color, linewidth=2, alpha=1)
        
            # axs[2].plot(sub['timestamp'], sub['calf_angle(degree)'], color=color, linewidth=2, alpha=1)
        
            # axs[3].plot(sub['timestamp'], sub['abs_knee_vel_avg(degree/s)'], color=color, linewidth=2, alpha=1)
    
    
    
        for ax in axs:
            ax.grid(True)
        axs[0].set_ylabel("Knee Angle (¬∞)")
        axs[1].set_ylabel("Thigh Angle (¬∞)")
        # axs[2].set_ylabel("Calf Angle (¬∞)")
        # axs[3].set_ylabel("Knee Velocity (¬∞/s)")
        axs[1].set_xlabel("Timestamp (s)")
        axs[0].legend()

        
        # Save plot
        plot_filename = f"{file_name}_gait_segment_{timestamp_str}.png"
        plot_path = os.path.join(save_folder, plot_filename)
        plt.tight_layout()
        plt.savefig(plot_path)
        plt.show()
        plt.close()
    
        print(f"üìä Saved plot: {plot_path}")

    # Label gait and phase
    gait_cycles, gait_labels, phase_labels, final_labels = label_gait_phases(filtered_df)

    final_labels = [phase if gait == 'Gait' else 'NonGait' for gait, phase in zip(gait_labels, phase_labels)]

    filtered_df['gait_label'] = gait_labels
    filtered_df['phase_label'] = phase_labels
    filtered_df['final_label'] = final_labels
      # Use timestamps as seconds relative to first timestamp (no extra *0.01)
    x_data = filtered_df['timestamp'] - filtered_df['timestamp'].iloc[0]
    # Phase durations
    labels = filtered_df['final_label'].tolist()
    segment_info = []
    start_idx = 0

    for i in range(1, len(labels)):
        if labels[i] != labels[i - 1]:
            seg_label = labels[i - 1]
            seg_start, seg_end = x_data.iloc[start_idx], x_data.iloc[i - 1]
            segment_info.append({
                'Phase': seg_label,
                'Start Time (s)': seg_start,
                'End Time (s)': seg_end,
                'Duration (s)': seg_end - seg_start
            })
            start_idx = i
    # Final segment
    segment_info.append({
        'Phase': labels[-1],
        'Start Time (s)': x_data.iloc[start_idx],
        'End Time (s)': x_data.iloc[-1],
        'Duration (s)': x_data.iloc[-1] - x_data.iloc[start_idx]
    })
    segment_df = pd.DataFrame(segment_info)

# Separate durations
    stance_durations = segment_df[segment_df['Phase'] == 'Stance']['Duration (s)'].reset_index(drop=True)
    swing_durations = segment_df[segment_df['Phase'] == 'Swing']['Duration (s)'].reset_index(drop=True)
    
    # Align lengths by padding shorter series with NaN
    max_len = max(len(stance_durations), len(swing_durations))
    stance_durations = stance_durations.reindex(range(max_len))
    swing_durations = swing_durations.reindex(range(max_len))
    
    # Compute Stance:Swing Ratio
    stance_swing_ratio = stance_durations / swing_durations

# 1. Compute Gait Cycle Durations
    gait_cycle_durations = []
    for start, end in gait_cycles:
        dur = filtered_df['timestamp'].iloc[end] - filtered_df['timestamp'].iloc[start]
        gait_cycle_durations.append(dur)

    # Reindex to match max_len (same length as stance/swing durations)
    gait_cycle_durations = pd.Series(gait_cycle_durations).reindex(range(max_len))
# 2. Combine into full duration table (before filtering)
    duration_table = pd.DataFrame({
        'Stance Duration (s)': stance_durations,
        'Swing Duration (s)': swing_durations,
        'Stance:Swing Ratio': stance_swing_ratio,
        'Gait Cycle Duration (s)': gait_cycle_durations
    })
# 3. Define a filter mask for valid durations
# how were filter values determined? - based on a mean ¬± 3œÉ of "duration table" 
# STANCE_MIN, STANCE_MAX = 0.2, 2.0
# SWING_MIN, SWING_MAX = 0.2, 1.5
# RATIO_MIN, RATIO_MAX = 0.5, 3.0
# GCD_MIN, GCD_MAX = 0.5, 3.0

    mask = (
        (duration_table['Stance Duration (s)'] >= 0.2) & 
        (duration_table['Stance Duration (s)'] <= 2) &
        (duration_table['Swing Duration (s)'] >= 0.2) &
        (duration_table['Swing Duration (s)'] <= 1.5) &
        (duration_table['Stance:Swing Ratio'] >= 0.5) &
        (duration_table['Stance:Swing Ratio'] <= 3) &
        (duration_table['Gait Cycle Duration (s)'] >= 0.5) &
        (duration_table['Gait Cycle Duration (s)'] <= 3)
    )

# 4. Apply the filter
    filtered_duration_table = duration_table[mask].reset_index(drop=True)
    gait_duration_df = pd.DataFrame({'Gait Cycle Duration (s)': gait_cycle_durations})
    filtered_gait_cycle_durations = duration_table.loc[mask, 'Gait Cycle Duration (s)'].reset_index(drop=True)

    idx = duration_table['Gait Cycle Duration (s)'].idxmax()
    #Show filtered cycles - could be used for further analysis of root causes
    print(filtered_df.iloc[gait_cycles[idx][0]:gait_cycles[idx][1] + 1])

    # Total duration of gait segments only
    gait_only_duration = sum(
        filtered_df['timestamp'].iloc[end] - filtered_df['timestamp'].iloc[start]
        for start, end in gait_cycles
    )
    gait_mins, gait_secs = divmod(gait_only_duration, 60)
    formatted_gait_duration = f"{int(gait_mins)} min {int(gait_secs)} sec"
    
    print(f"üö∂‚Äç‚ôÇÔ∏è Total Gait-Only Duration: {gait_only_duration:.2f} seconds ({formatted_gait_duration})")





    # Display and optionally save
    print("\nüìä Saved Phase Duration Summary:")
    print(duration_table)
    print("\n‚úÖ Saved Filtered Valid Durations:")
    print(filtered_duration_table)
#step count
    step_count = len(filtered_duration_table)
    print(f"\nüö∂ Step Count (Valid Gait Cycles): {step_count}")

    # Add Step Count to the last row only
    filtered_duration_table['Step Count'] = None
    if not filtered_duration_table.empty:
        filtered_duration_table.at[filtered_duration_table.index[-1], 'Step Count'] = step_count

#save durations
    duration_table.to_csv(os.path.join(save_folder, "all_durations.csv"), index=False)
    filtered_duration_table.to_csv(os.path.join(save_folder, "filtered_durations.csv"), index=False)

    #stats analysis- unfiltered and filtered data 
subject = extract_subject(file_name)
if not filtered_duration_table.empty:
    stats = filtered_duration_table.describe().transpose()
    stats = stats.reset_index().rename(columns={"index": "Metric"})
    stats["Subject"] = subject
    stats["File"] = file_name
    stats["Trial"] = file_name.split("_algorithm")[0]  # or just file_name.split('.')[0]
    summary_stats.append(stats)

    print("üßæ Columns in all_trial_stats:", all_trial_stats.columns.tolist())
    # Save per-trial summary inside its folder
    summary_path = os.path.join(save_folder, "trialwise_stats_summary.csv")
    stats.to_csv(summary_path, index=False)

 

import seaborn as sns
import matplotlib.pyplot as plt
if summary_stats:
    all_trial_stats = pd.concat(summary_stats, ignore_index=True)

    print("\nüìä Saved Summary Statistics by Trial:")
    print(all_trial_stats)

    # ‚úÖ Prepare data for plotting
    plot_df = all_trial_stats[all_trial_stats['Metric'].isin([
        'Stance Duration (s)', 'Swing Duration (s)', 'Gait Cycle Duration (s)', 'Stance:Swing Ratio'
    ])]

    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=plot_df,
        x="Trial", y="mean", hue="Metric", palette="Set2"
    )
    plt.title("Mean Gait Metrics per Trial")
    plt.ylabel("Mean Duration (seconds or ratio)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```
